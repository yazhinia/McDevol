# distance from read + kmer counts (no Bayesian Factor)

```python
labels = pd.read_csv("/big/work/metadevol/benchmark_dataset1/contigs_umaplabel", header=None, sep='\t')
labels.columns = ["contig_id","contig_name","length","genome_id","connected_comp","bin"]
```

```python
cdist_matrix = np.load("/big/work/metadevol/benchmark_dataset1/cdist_newcheck_matrix.npy", allow_pickle=True)
my_model_comb = umap.UMAP(metric='precomputed')
my_model_fit_comb = my_model_comb.fit(np.matrix(cdist_matrix))
umap.plot.points(my_model_fit_comb, labels=labels['connected_comp'], color_key=colors)
```

```python
umap.plot.points(my_model_fit_comb, labels=labels['bin'], color_key=colors)
# plt.savefig("/big/work/metadevol/benchmark_dataset1/umap_coloredbygenomeid.png",dpi=600,format="png",bbox_inches="tight")
```

```python
umap.plot.output_notebook()
pp = umap.plot.interactive(my_model_fit_comb, labels=labels["connected_comp"], hover_data=labels, theme='fire', point_size=2)
umap.plot.show(pp)
```
