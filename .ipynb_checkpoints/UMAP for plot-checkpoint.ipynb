```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
import umap
import umap.plot




```

```python
tmp_dir = "/big/work/metadevol/new_simdata/eachsamreads_allcontigs_allalign/tmp/"
labels = pd.read_csv(tmp_dir + 'contigs_refid_genomeid', header=None, sep=' ')
labels.columns = ["index","contig_name","genome_id","length"]
```

```python
read_matrix = np.load(tmp_dir + "X_metadevoldistreadwithcap.npy", allow_pickle=True)
my_model_read = umap.UMAP(metric='precomputed', low_memory=True)
my_model_fit_read = my_model_read.fit(np.matrix(read_matrix))
```

```python
umap.plot.points(my_model_fit_read)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
umap.plot.connectivity(my_model_fit_read, color_key_cmap='grey') # with repeat correction + capping
plt.savefig(tmp_dir+'connected_component.png', dpi=600, bbox_inches='tight', format='png')
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
fig, ax = plt.subplots(figsize=(4,4))
umap_points = pd.DataFrame(np.vstack([my_model_fit_read.embedding_[:,0],my_model_fit_read.embedding_[:,1]]).T)
plot_data = umap_points.merge(labels, left_index=True, right_index=True)
cluster_id = pd.read_csv(tmp_dir + 'cluster_assigned_onlyreadsq8', header=None)
plot_data = plot_data.merge(cluster_id, left_index=True, right_index=True)
plot_data.columns = [0,1,'index','contig','genome_id', 'length','cluster_id']
# colors = {'17_36':'red', '17_14':'maroon', '17_0':'orange', '4_0':'blue', '4_1':'green'}
# ax.scatter(plot_data[0],plot_data[1], c=plot_data['genome_id'].map(colors), s=1, alpha=0.6)
# g=sns.scatterplot(plot_data[0], plot_data[1], data=plot_data, hue=plot_data['genome_id'].map(colors), linewidth=0, s=1,legend=False)
# g=sns.scatterplot(plot_data[0], plot_data[1], data=plot_data, hue='cluster_id', linewidth=0, s=1, palette="deep", legend=False)
# plt.xlabel('umap embedding axis 1')
# plt.ylabel('umap embedding axis 2')
# sns.move_legend(g, "upper left", bbox_to_anchor=(.55, .45), title='genome id')
# plt.savefig(tmp_dir+'readdistclusterid_umap.png', dpi=600, bbox_inches='tight', format='png')


# g=sns.scatterplot(plot_data[0], plot_data[1], data=plot_data, linewidth=0, s=1, palette='gray', legend=False)
ax.scatter(plot_data[0],plot_data[1], c='gray', s=0.5)
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.yticks([])
plt.savefig(tmp_dir+'clustering_sample.png', dpi=600, bbox_inches='tight', format='png')

```
